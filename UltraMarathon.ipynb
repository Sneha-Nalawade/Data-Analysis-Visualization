now, lets see the data specifically about year 2020, where event/track distance was either 50 km or 50 miles

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)]
```

now, lets extract the data for only one event, and that is: Everglades 50 Mile Ultra Run (USA)

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

alternate way to achieve the above result:

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

combining the above two sets of conditions:

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(5)
```

```python
df2.shape
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)     #we're removing USA since all the events in df2 are anyways belonging to USA, so its just redundant and not necessary
```

```python
df2.head(5)
```

getting athlete ages for all rows, and storing it in a new column

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(4)
```

drop the unnecessary columns from df2

```python
df2 = df2.drop(['Athlete club', 'Athlete country','Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(4)
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

remove rows/entries/records with null value in any of the fields

```python
df2 = df2.dropna()
```

```python
df2.shape
```

check for duplicates

```python
df2[df2.duplicated() == True]  # No duplicates found/exist
```

```python
df2.head(3)
```

```python
df2.reset_index(drop=True)
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns={'Year of event':'year',
                          'Event dates':'race_day',
                          'Event name':'race_name',
                          'Event distance/length':'race_length',
                          'Event number of finishers':'race_number_of_finishers',
                          'Athlete performance':'athlete_performance',
                          'Athlete gender':'athlete_gender',
                          'Athlete average speed':'athlete_average_speed',
                          'Athlete ID':'athlete_id'})
```

what age groups are best in the 50 mi Race (20 + races min) (show 15)

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```

what age groups are the worst in the 50m race (10 + races min) (show 20)

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>9').head(20)
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2.head(3)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

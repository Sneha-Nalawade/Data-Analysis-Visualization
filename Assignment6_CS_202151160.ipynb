Importing the necessary Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Reading Data:

```python
female_population = pd.read_csv('/content/drive/MyDrive/DAV sem 6/population_of_female - population_of_female.csv')
female_completion = pd.read_csv('/content/drive/MyDrive/DAV sem 6/primary_completion_rate_female - primary_completion_rate_female.csv')
```

1. Create a column named “average_rate” in “primary_completion_rate_female.csv” that contains the average (over years) completion rate for each country.

```python
female_completion['avgRate'] = female_completion.mean(axis=1, numeric_only=True)
df1 = female_completion[['Country Name', 'avgRate']]
df1.head()
```

2. Create a column named “Latest Population” in “population_of_female.csv” that contains the latest population of females in the country.

```python
female_population['Latest Population'] = female_population.iloc[:,-1]
female_population.head()
```

3. What is the weighted average of “average_rate”? The weight should be calculated based on “latest population” of females in the country.

```python
wtAvg, sum = 0, 0
for i in range(len(female_completion)):
    if np.isnan(female_completion['avgRate'][i]):
        continue
    else:
        wtAvg += female_completion['avgRate'][i] * female_population['Latest Population'][i]
        sum += female_population['Latest Population'][i]
wtAvg /= sum
print(wtAvg)
```

4. What is the Median of “average_rate”?

```python
median = female_completion['avgRate'].median()
print(median)
```

5. What is the InterQuartile Range of “average_rate”?

```python
q1, q3 = female_completion['avgRate'].quantile([.25,.75])
iqr = q3-q1
print(iqr)
```

6. What is the Variance of “average_rate”?

```python
variance = female_completion['avgRate'].var()
print(variance)
```
